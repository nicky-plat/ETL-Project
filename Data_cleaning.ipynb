### Loading csv files

```python
# load csv into dataframe

# sales files
coke_sales = pd.read_csv('data/Coke_data_cleaned.csv')
pepsi_sales = pd.read_csv('data/Pepsi_Receipts_Report_CLEAN.csv')
jj_sales = pd.read_csv('data/Jack&Jill_19-20_Receipts Report_.CLEAN.csv')

# list files
coke_list = pd.read_csv('data/Coke_items.csv')
pepsi_list = pd.read_csv('data/pepsi_items.csv')
jj_list = pd.read_csv('data/jj_items.csv')
```

### Cleaning up Sales files

```python
# Renaming Columns
coke_sales = coke_sales.rename(columns={
    'Item Name\n': 'Item_Name',
    'Size\n': 'Size',
    'Order\nQty': 'Order_Qty',
    'Rcvd\nQty': 'Rcvd_Qty',
    'Cost\n': 'Cost',
    'Extended\nCost': 'Extended_Cost'
})
coke_sales = coke_sales[['Item_Name','Size','Order_Qty','Rcvd_Qty','Cost','Extended_Cost']]

# Renaming Columns
coke_sales = coke_sales.dropna(how='any')

coke_sales.head()
```

```python
# Renaming Columns
pepsi_sales = pepsi_sales.rename(columns={
    'Item Name\n': 'Item_Name',
    'Size\n': 'Size',
    'Order\nQty': 'Order_Qty',
    'Rcvd\nQty': 'Rcvd_Qty',
    'Cost\n': 'Cost',
    'Extended\nCost': 'Extended_Cost'
})

# Renaming Columns
pepsi_sales = pepsi_sales.dropna(how='any')

pepsi_sales
```

```python
jj_sales = jj_sales.rename(columns={
    'Item Name\n': 'Item_Name',
    'Size\n': 'Size',
    'Order\nQty': 'Order_Qty',
    'Rcvd\nQty': 'Rcvd_Qty',
    'Cost\n': 'Cost',
    'Extended\nCost': 'Extended_Cost'
})
jj_sales = jj_sales[['Item_Name','Size','Order_Qty','Rcvd_Qty','Cost','Extended_Cost']]
jj_sales = jj_sales.dropna(how='any')
jj_sales
```
